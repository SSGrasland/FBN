```python
df.dropna(subset = ['Vol', 'No', 'Year'], inplace=True)
```

```python
df[['No', 'Page Start', 'Page End']] = df[['No', 'Page Start', 'Page End']].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
df['Issue Length'] = df.groupby(['Vol', 'No'])['Page End'].transform(max)
```

```python
df
```

```python
df['ArticleLength'] = (df['Page End'] - df['Page Start']) + 1
df.head()
```

```python
df['ArticlePercent'] = (df['ArticleLength']/df['Issue Length'])*100
```

```python
df.head()
```

```python
df['ArticleTitle'] = df['Article Title']
df['PageStart'] = df['Page Start'] 
df['PageEnd'] = df['Page End']
df['IssueLength'] = df['Issue Length']
df['ContributorName'] = df['Contributor Name']
df['Issue'] = df['No']
df.head()
```

```python
df = df[['ContributorName', 'Vol', 'No', 'Year', 'Month', 'ArticleTitle', 'PageStart', 'PageEnd', 'ArticleLength', 'ArticlePercent', 'IssueLength', 'Author', 'Category', 'Graphics?', 'Notes', 'File']]
df.head()
```
