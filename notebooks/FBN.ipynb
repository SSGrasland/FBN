```python
df['Issue Length'] = df.groupby(['Vol', 'No'])['PageEnd'].transform(max)
```

```python
df['ArticleLength'] = (df['PageEnd'] - df['PageStart']) + 1
df.head()
```

```python
df['ArticlePercent'] = (df['ArticleLength']/df['IssueLength'])*100
```

```python
df = df[['ContributorName', 'Vol', 'No', 'Year', 'Month', 'ArticleTitle', 'PageStart', 'PageEnd', 'ArticleLength', 'ArticlePercent', 'IssueLength', 'Author', 'Category', 'Graphics?', 'Notes', 'File']]
df.head()
```

```python
df.Month.unique()
```

```python
#Unifying Month Values 
row_labels = {'(Sep/Oct?)':'September','May-June':'May', 'Dec':'December',
       'Feb':'February', 'Augusr':'August', 'Nov/Dec.':'November', 'Nov':'November', 'Jan/Feb':'January', 'Mar/Apr':'March',
       'May/June':'May', 'Sept / Oct':'September', 'Sept/Oct':'September', 'March/April':'March',
       'November/December':'November'}

df['Month']=df['Month'].replace(row_labels)
```

```python
row_labels = {'Other Feature' : 'Other', 'Letter ':'Letters', 'Booklist/Media list' : 'Booklist/Medialist', 'Booklist':'Booklist/Medialist'}

df['Category']=df['Category'].replace(row_labels)
```
