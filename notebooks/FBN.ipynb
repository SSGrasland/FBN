## Joining Datasets

```python
#renaming columns so we can join datasets 

df.rename(columns={"Vol": "Volume", "No": "Issue"}, inplace= True)
date.rename(columns={"Volume #": "Volume", "Issue #": "Issue"}, inplace =True)
```

```python
#any issue that doesn't have a Date Ended means it's incomplete
#dropping any incomplete issue 
date.dropna(subset = ['Date Ended'], inplace= True)
date
```

```python
df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')
df['Issue'] = pd.to_numeric(df['Issue'], errors='coerce')
date['Volume'] = pd.to_numeric(date['Volume'], errors='coerce')
date['Issue'] = pd.to_numeric(date['Issue'], errors='coerce')
date['Duration (In Days)'] = pd.to_numeric(date['Duration (In Days)'], errors='coerce')
```

```python
date['Person Working on it'].value_counts()
```

```python
date = date[['Volume', 'Issue', 'Date Started', 'Date Ended', 'Duration (In Days)']]
date.head()
```

```python
#joining the datasets date and df so we can drop any info from df that is incomplete
#setting the index to Volume and Issue so that we use .join()
df.set_index(['Volume', 'Issue'], inplace = True)
date.set_index(['Volume', 'Issue'], inplace = True)
```

```python
#joining df and date datasets
df = df.join(date)
```

```python
#reseting the index
df.reset_index(inplace = True)
df.head()
```

```python
#any issue that doesn't have a Date Ended means it's incomplete
#dropping any incomplete issue 
df.dropna(subset = ['Date Ended'], inplace= True)
df
```

```python
#dropping all 'Unnamed' columns
# Create a boolean array indicating which columns contain the string "Unnamed"
cols_to_drop = df.columns[df.columns.str.contains('Unnamed')]

# Drop the columns containing the string "Email"
df.drop(cols_to_drop, axis=1, inplace=True)
df
```

```python
df[['Page Start', 'Page End', 'Volume']] = df[['Page Start', 'Page End', 'Volume']].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
df['Issue Length'] = df.groupby(['Volume', 'Issue'])['Page End'].transform(max)
```

```python
df
```

## Data Engineering

### Column for Article Length

```python
df['Article Length'] = (df['Page End'] - df['Page Start']) + 1
df.head()
```

```python
df['Article Percent'] = (df['Article Length']/df['Issue Length'])*100
```

```python
df.head()
```

```python
#reordering columns
df = df[['Contributor Name', 'Volume', 'Issue', 'Year', 'Month', 'Article Title', 'Page Start', 'Page End', 'Article Length', 'Article Percent', 'Issue Length', 'Date Started', 'Date Ended', 'Duration (In Days)', 'Author', 'Category', 'Graphics?', 'Notes', 'File']]
df.head()
```

## Column For Time it Takes To Index One Page 

```python
df['Time To Index One Page'] = df['Issue Length']/df['Duration (In Days)']
```

# Data Cleaning

## Making Input Uniform 

### Month

```python
df.Month.unique()
```

```python
#Unifying Month Values 
row_labels = {'(Sep/Oct?)':'September','May-June':'May', 'Dec':'December',
       'Feb':'February', 'Augusr':'August', 'Nov/Dec.':'November', 'Nov':'November', 'Jan/Feb':'January', 'Mar/Apr':'March',
       'May/June':'May', 'Sept / Oct':'September', 'Sept/Oct':'September', 'March/April':'March',
       'November/December':'November'}

df['Month']=df['Month'].replace(row_labels)
```

```python
df.Month.unique()
```

### Category

```python
df.Category.unique()
```

```python
row_labels = {'Other Feature' : 'Other', 'Letter ':'Letters', 'Booklist/Media list' : 'Booklist/Medialist', 'Booklist':'Booklist/Medialist'}

df['Category']=df['Category'].replace(row_labels)
```
